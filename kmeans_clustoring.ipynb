```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris=load_iris()
```

```python


# Check the shape of the data
print(iris.data.shape)

# Change the number of columns in the data to match the number of column names
iris.data = iris.data[:, :2]

# Create the DataFrame
iris = pd.DataFrame(iris.data, columns=['sepal length (cm)', 'sepal width (cm)'])

# Print the DataFrame
print(iris)
```

```python
plt.scatter((iris['sepal length (cm)']),(iris['sepal width (cm)']))
```

```python
iris.values
```

```python
from sklearn.cluster import  KMeans
km=KMeans(n_clusters=2 )
y_predicted =km.fit_predict(iris)
```

```python
km.predict([[5.1,3.5]])
```

```python
iris["cluster"]= y_predicted
```

```python
iris
```

```python
plt.scatter((iris['sepal length (cm)']),(iris['sepal width (cm)']))
iris1=iris[iris.cluster==0]
iris2=iris[iris.cluster==1]
plt.scatter((iris1['sepal length (cm)']),(iris1['sepal width (cm)']),color='g')
plt.scatter((iris2['sepal length (cm)']),(iris2['sepal width (cm)']),color='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker="*",color="y")
```
